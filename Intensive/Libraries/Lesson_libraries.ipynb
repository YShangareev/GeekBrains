### Matplotlib/Seaborn


* Нарисуйте график sin(x), подпишите оси, добавьте сетку
* Нарисуйте график sin(x) и cos(x) на одном subplot'е и на разных, задайте цвета, которые вам больше нравятся

```python
import matplotlib.pyplot as plt
# Создание массива от -пи до пи, с шагом 0.01(чем меньше шаг, тем плавнее линия)
x = np.arange(-np.pi, np.pi, 0.01)
# Определяем функцию
y = np.sin(x)
y1 = np.cos(x)
# Создаем фигуру
plt.figure()
# График по точкам х и у
plt.plot(x, y, '-', color='y')
plt.plot(x, y1, '--', color='g')
# Добавляем сетку
plt.grid()
# Подписываем график, оси, и функции
plt.title('График sin x и cos x')
plt.ylabel('Ось y')
plt.xlabel('Ось x')
plt.legend(labels=['sin x', 'cos x'])

plt.show()

# Создаем область figure с размерами
plt.figure(figsize=(16, 16))
# Условно разделили нашу область на 2 части и добавили график
plt.subplot(2, 1, 1)
plt.plot(x, y, '-', color='y')
plt.grid()
plt.title('График sin x')
plt.ylabel('Ось y')
plt.xlabel('Ось x')

plt.subplot(2, 1, 2)
plt.plot(x, y1, '--', color='g')
plt.grid()
plt.title('График cos x')
plt.ylabel('Ось y')
plt.xlabel('Ось x')
plt.show()
```
