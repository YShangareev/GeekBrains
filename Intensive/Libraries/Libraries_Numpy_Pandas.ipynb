```python
import numpy as np

A_mat = np.array([[1, -2, 4],
                  [2, 0, -1]])
B_mat = np.array([[5, 2, 3],
                  [4, 6, 2]])
print(f'{A_mat}\n{B_mat}')
print(A_mat - B_mat)
print(A_mat + B_mat)

A = np.array([[1, 2],
              [2, -1],
              [3, 0]])
B = np.array([[-1, 1],
              [1, 2],
              [0, 0]])
C = A - 3 * B
print(C)

# Задача 3. Должны быть матрицы одного размера

a = ([[6, -1, 4],
      [2, 1, 10],
      [8, 0, 2]])
print(np.linalg.det(a))

b = ([[3, 3, -1],
      [4, 1, 3],
      [1, -2, -2]])
c = np.linalg.inv(b)
d = np.array(np.linalg.inv(c), dtype=int)

print(np.round(c, 2))
print(d)
```
