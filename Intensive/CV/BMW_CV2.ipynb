```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('bmw_3.jpg', cv2.IMREAD_COLOR)
```

```python
plt.imshow(img)
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img)
```

```python
RED = (255, 0, 0)
BLUE = (0, 0, 255)

p0 = 50, 200
p1 = 700, 500
```

```python
cv2.rectangle(rgb_img, p0, p1, BLUE, 2)
plt.imshow(rgb_img)
```

```python
center = 400, 350
axes = 400, 200
angle = 0
```

```python
rgb_img_1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.ellipse(rgb_img_1, center, axes, angle, 0, 360, RED, 5)
plt.imshow(rgb_img_1)
```

```python
gray_img = cv2.imread('bmw_3.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(gray_img, cmap = 'gray')
```

```python
img_2 = cv2.GaussianBlur(gray_img, (9, 9), 0)
plt.imshow(img_2, cmap = 'gray')
```

```python
img_2 = cv2.GaussianBlur(gray_img, (21, 21), 0)
plt.imshow(img_2, cmap = 'gray')
```

```python
x = cv2.Sobel(gray_img, cv2.CV_16S, 0, 1)
abs_x = cv2.convertScaleAbs(x)
plt.imshow(abs_x, cmap = 'gray')
```

```python
x = cv2.Sobel(gray_img, cv2.CV_16S, 1, 0)
abs_x = cv2.convertScaleAbs(x)
plt.imshow(abs_x, cmap = 'gray')
```
